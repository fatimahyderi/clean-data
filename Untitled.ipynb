```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
tb = pd.read_csv('tb.csv')
```

```python
tb.head()
```

```python
tb.columns
```

```python
tb=tb.melt(id_vars=["country","year"],var_name=["sexage"],value_name="cases")
```

```python
tb.head()
```

```python
tb=(tb.assign(
sex = lambda x: x.sexage.str[0].astype(str),
age = lambda x: x.sexage.str[1:].astype(str))
      .drop("sexage",axis=1))
tb
```

```python
tb=(tb.replace(to_replace =["04","514","2534","4554","3544","5564","1524","65","u"],value =["0-4","5-14","25-34","45-54","35-44","55-64","15-24","65+",np.nan])
           )
```

```python
tb
```

```python
tb=tb[["country","year","age","sex","cases"]]
```

```python
tb
```

```python
tb.dropna(subset = ["cases"])
```

```python
tb.to_csv('final.csv',index=False)
```
